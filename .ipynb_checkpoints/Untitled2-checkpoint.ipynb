```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
from pandas.plotting import autocorrelation_plot
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
import warnings
from statsmodels.tools.sm_exceptions import ValueWarning
warnings.simplefilter('ignore', ValueWarning)
```

```python
data = pd.read_csv('UCL Players Modified.csv', encoding='ISO-8859-1')
```

```python
# Convertir la colonne de date en type de données datetime
data['Date_Time'] = pd.to_datetime(data['Date_Time'])
```

```python
# Trier les données par ordre chronologique
data = data.sort_values(by='Date_Time')
```

```python
# Afficher l'évolution des buts marqués en fonction du temps
plt.figure(figsize=(12,6))
sns.lineplot(data=data, x='Date_Time', y='Gls')
plt.title('Évolution des buts marqués en fonction du temps')
plt.xlabel('Date_Time')
plt.ylabel('Nombre de buts marqués')
plt.show()
```

```python
#Le premier graphique montre l'évolution des buts marqués en fonction du temps. 
#On peut observer que le nombre de buts marqués varie considérablement au fil du temps, avec des périodes de hausses et de baisses.
```

```python
# Afficher l'évolution des tirs cadrés en fonction du temps
plt.figure(figsize=(12,6))
sns.lineplot(data=data, x='Date_Time', y='SoT')
plt.title('Évolution des tirs cadrés en fonction du temps')
plt.xlabel('Date_Time')
plt.ylabel('Nombre de tirs cadrés')
plt.show()
```

```python
#Le deuxième graphique montre l'évolution des tirs cadrés en fonction du temps. 
#On peut observer que le nombre de tirs cadrés fluctue également au fil du temps.
```

```python
#Décomposition des séries temporelles
```

```python
# Décomposition des buts marqués
result_gls = seasonal_decompose(data['Gls'], model='additive', period=30)
plt.figure(figsize=(12,8))
result_gls.plot()
plt.show()
```

```python
# Décomposition des tirs cadrés
result_sot = seasonal_decompose(data['SoT'], model='additive', period=30)
plt.figure(figsize=(12,8))
result_sot.plot()
plt.show()
```

```python
#Verification de stationnarité
```

```python
# Statistiques mobiles
rolmean = data['Gls'].rolling(window=30).mean()
rolstd = data['Gls'].rolling(window=30).std()
```

```python
# Tracer les statistiques mobiles
plt.figure(figsize=(12,6))
plt.plot(data['Gls'], color='blue', label='Original')
plt.plot(rolmean, color='red', label='Moyenne mobile')
plt.plot(rolstd, color='black', label='Écart-type mobile')
plt.legend(loc='best')
plt.title('Moyenne et écart-type mobiles')
plt.show()
```

```python
#L'hypothèse nulle du test ADF est que la série temporelle a une unité racine, c'est-à-dire qu'elle est non stationnaire. Si la valeur des statistiques ADF est inférieure à la valeur critique à un niveau de confiance donné, on peut rejeter l'hypothèse nulle.
#Dans ce cas, puisque la valeur des statistiques ADF (-12.97) est inférieure aux valeurs critiques à tous les niveaux de confiance, et que la valeur p est extrêmement faible, nous pouvons rejeter l'hypothèse nulle.
#Par conséquent, la série temporelle est stationnaire.
```

```python
# Autocorrelation Function (ACF)
plt.figure(figsize=(12, 6))
plot_acf(data['Gls'], lags=30, ax=plt.gca())
plt.title('Autocorrelation Function (ACF)')
plt.show()
```

```python
# Partial Autocorrelation Function (PACF)
plt.figure(figsize=(12, 6))
plot_pacf(data['Gls'], lags=30, ax=plt.gca())
plt.title('Partial Autocorrelation Function (PACF)')
plt.show()
```

```python
# Fonctions d'autocorrélation (ACF) et d'autocorrélation partielle (PACF)
plt.figure(figsize=(12,6))
plt.subplot(121)
plot_acf(data['Gls'], ax=plt.gca(), title="Autocorrélation pour Gls")
plt.subplot(122)
plot_pacf(data['Gls'], ax=plt.gca(), title="Autocorrélation partielle pour Gls")
plt.tight_layout()
plt.show()
```

```python
model = ARIMA(data['Gls'], order=(1,1,1)) # Les valeurs (1,1,1) sont à titre d'exemple
model_fit = model.fit()
forecast = model_fit.forecast(steps=30) # Prévoir les 30 prochains points
```

```python
# Afficher la prévision
plt.figure(figsize=(12,6))
plt.plot(data.index, data['Gls'], label='Données observées')
plt.plot(pd.date_range(data.index[-1], periods=31, closed='right'), forecast, label='Prévision', color='red')
plt.title("Prévision ARIMA")
plt.legend()
plt.show()
```
